## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
bins = [0,9.9,14,19,24,29,34,39,100]
group_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)
purchase_data_df.head()

# purchase_data_df = purchase_data_df.groupby("Age")
```

```python
# This gives total counts (does not give unique counts)
purchase_data_df['Age Range'].value_counts()
```

```python
grouped_age_df = purchase_data_df.groupby("Age Range")
price_agerange_df = pd.DataFrame(grouped_age_df["Price"].sum())
price_agerange_df
```

```python
unique_valuesperage_df=grouped_age_df.nunique()
unique_valuesperage_df=unique_valuesperage_df[["Purchase ID", "SN"]]
unique_valuesperage_df
```

```python
merge_df=pd.merge(unique_valuesperage_df,price_agerange_df,on="Age Range")
merge_df
```

```python
renamed_df=merge_df.rename(columns=
            {"Purchase ID":"Purchase Count","SN":"Unique Purchasers","Price":"Total Purchase Value"})
renamed_df = renamed_df[["Total Purchase Value","Purchase Count","Unique Purchasers"]]
renamed_df
```

```python
price_per_purchase = renamed_df["Total Purchase Value"]/renamed_df["Purchase Count"]
spending_per_user = renamed_df["Total Purchase Value"]/renamed_df["Unique Purchasers"]
renamed_df["Average Purch Price"]= price_per_purchase
renamed_df["Avg Total Purchase per Person"]= spending_per_user 
```

```python
percentage_players=renamed_df["Unique Purchasers"]/unique_player_count
renamed_df["Percentage of Players"]= percentage_players
renamed_df3 = renamed_df[["Unique Purchasers","Percentage of Players"]]
renamed_df3
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data_df.head()
```

```python
itemname_df=purchase_data_df[['Item ID']]
itemname_df=itemname_df.rename(columns={'Item ID':"Item ID2"})
itemname_df.head()
```

```python
purchase_data_df = pd.concat([itemname_df, purchase_data_df], axis=1)
purchase_data_df.head()
```

```python
grouped_item_df = purchase_data_df.groupby(["Item ID","Item Name"])

item_arrange_df = pd.DataFrame(grouped_item_df["Item ID2"].count()).sort_values("Item ID2",ascending=False)

item_arrange_df = item_arrange_df.rename(columns= {
                    "Item ID2":"Purchase Count"})
item_arrange_df.head()
```

```python
item_arrange2_df = pd.DataFrame(grouped_item_df["Price"].sum()).sort_values("Price",ascending=False)
item_arrange2_df = item_arrange2_df.rename(columns= {
                    "Price":"Total Purchase Value"})
item_arrange2_df.head()
```

```python
biggest_by_volume_df=pd.merge(item_arrange_df,item_arrange2_df,on=['Item ID','Item Name']).sort_values("Purchase Count",ascending=False)
item_price=biggest_by_volume_df["Total Purchase Value"]/item_arrange_df["Purchase Count"]
biggest_by_volume_df["Item Price"]=item_price
biggest_by_volume_top5_df=biggest_by_volume_df.iloc[0:5,:]
biggest_by_volume_top5_df=biggest_by_volume_top5_df[["Purchase Count", "Item Price","Total Purchase Value"]]
biggest_by_volume_top5_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
biggest_by_total_value_df=pd.merge(item_arrange_df,item_arrange2_df,on=['Item ID','Item Name']).sort_values("Total Purchase Value",ascending=False)
item_price=biggest_by_total_value_df["Total Purchase Value"]/item_arrange_df["Purchase Count"]
biggest_by_total_value_df["Item Price"]=item_price
bigget_by_total_value_top5_df=biggest_by_total_value_df.iloc[0:5,:]
bigget_by_total_value_top5_df=bigget_by_total_value_top5_df[['Purchase Count','Item Price','Total Purchase Value']]
bigget_by_total_value_top5_df
```
